```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import Imputer
from sklearn.ensemble import RandomForestClassifier
```

```python
lyrics = pd.read_csv('lyrics.csv')
```

```python
lyrics
```

```python
holdout = pd.read_csv('holdout_attributes.csv')
```

```python
holdout
```

```python
rf = RandomForestClassifier(n_estimators=1000, n_jobs=-1)
```

```python
labels = ['Hip-hop', 'Country']
```

```python
mapped_labels = list(map(lambda x: labels.index(x), lyrics.Genre.values))
```

```python
rf.fit(lyrics.loc[:, 'i':], mapped_labels)
```

```python
lyrics.loc[:, 'i':]
```

```python
i = 0
for column in lyrics.loc[:, 'i':]:
    
    if column not in holdout.loc[:, 'i':]:
        holdout.insert(i + 1, column, [float('NaN')] * 100)
        
    i += 1
holdout.loc[:, 'i':]
```

```python
holdout
```

```python
holdout[holdout.isnull()] = 0
holdout
```

```python
predictions = list(map(lambda x: labels[x], rf.predict(holdout.loc[:, 'i':])))

submission = pd.DataFrame()
submission['Id'] = np.arange(100)
submission['Predictions'] = predictions
submission.to_csv('submission.csv', index=False)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNC = KNeighborsClassifier(n_neighbors=5, n_jobs=-1)
```

```python
KNC.fit(lyrics.loc[:, 'i':], mapped_labels)
```
